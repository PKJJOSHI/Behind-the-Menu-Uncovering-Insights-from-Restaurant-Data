```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load the dataset
df = pd.read_csv('Dataset.csv')
```

```python
# Identify restaurant chains by checking for duplicate names
# Group by 'Restaurant Name' and count occurrences
restaurant_chains = df['Restaurant Name'].value_counts()
chains = restaurant_chains[restaurant_chains > 1]
print(f"Identified {len(chains)} restaurant chains.")
print("\nRestaurant Chains and their counts:")
print(chains)
```

```python
# Filter the dataset to include only restaurant chains
chains_df = df[df['Restaurant Name'].isin(chains.index)]
```

```python
# Analyze ratings of different restaurant chains
average_ratings = chains_df.groupby('Restaurant Name')['Aggregate rating'].mean().sort_values(ascending=False)
print("\nAverage Ratings of Restaurant Chains:")
print(average_ratings)
```

```python
# Analyze popularity of restaurant chains (using 'Votes' as a metric)
average_votes = chains_df.groupby('Restaurant Name')['Votes'].mean().sort_values(ascending=False)
print("\nAverage Votes (Popularity) of Restaurant Chains:")
print(average_votes)
```

```python
# Plot the top 10 restaurant chains by average rating
plt.figure(figsize=(12, 8))
sns.barplot(x=average_ratings.head(10), y=average_ratings.head(10).index, palette="magma")
plt.title('Top 10 Restaurant Chains by Average Rating')
plt.xlabel('Average Aggregate Rating')
plt.ylabel('Restaurant Chain')
plt.show()
```

```python
# Plot the top 10 restaurant chains by popularity (average votes)
plt.figure(figsize=(12, 8))
sns.barplot(x=average_votes.head(10), y=average_votes.head(10).index, palette="viridis")
plt.title('Top 10 Restaurant Chains by Popularity (Average Votes)')
plt.xlabel('Average Votes')
plt.ylabel('Restaurant Chain')
plt.show()
```
